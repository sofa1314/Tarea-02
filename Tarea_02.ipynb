```python
#Problema 1
lista_numeros = range(-30,51)

for x in lista_numeros:
  if x < 0:
    print(f"{x}: Muy frío")
  elif 0 <= x < 25:
    print(f"{x}: Templado")
  else:
    print(f"{x}: Caluroso")
```

```python
#Problema 2
#Números especiales del robot

for numero in range(1, 101):
  if numero % 5 == 0:
    if numero % 10 == 0:
      print(f"{numero} Bip Bip")
    else:
      print(f"{numero} Bip")
```

```python
#Problema 3
#Detector de puntaje gamer

num1 = float(input("Ingresa primer puntaje: "))
num2 = float(input("Ingresa segundo puntaje: "))

suma = num1 + num2

print(f"Resultado: ", end="")
if suma < 50:
  print("Nivel principiante")
elif 50 <= suma < 80:
  print("Nivel intermedio")
else:
  print("Nivel pro")
```

```python
#Problema 4
#Calculador de perfil musical

edad = int(input("Ingresa tu edad: "))
le_gusta_musica_str = input("¿Te gusta la música? (Sí/No): ").lower()
le_gusta_bailar_str = input("¿Te gusta bailar? (Sí/No): ").lower()

le_gusta_musica = le_gusta_musica_str == 'sí' or le_gusta_musica_str == 'si'
le_gusta_bailar = le_gusta_bailar_str == 'sí' or le_gusta_bailar_str == 'si'

if edad >= 18 and le_gusta_musica and le_gusta_bailar:
  print("Eres una estrella de la fiesta")
elif le_gusta_musica and not le_gusta_bailar:
  print("Te gusta la música, pero no tanto bailar")
else:
  print("Tienes tu propio estilo")
```
